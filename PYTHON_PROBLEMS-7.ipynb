# Problem 1: Smart Light Controller

### Problem Statement

Create a class to control a smart light that can be turned ON or OFF and display its status.

### üåç Real-Time Use

Smart home automation systems

### üì• Sample Input
Light Name: Bedroom Light

Action: ON

### üì§ Expected Output
Bedroom Light is ON


```python
class SmartLightController:
    def __init__(self, action, light_name):
        self.action = action
        self.light_name = light_name

    def light(self):
        if self.action.lower() == "on":
            return f"{self.light_name} is ON"
        elif self.action.lower() == "off":
            return f"{self.light_name} is OFF"
        else:
            return "Invalid action"

name = input("Light name:")
acti = input("Action (on/off):")

smart = SmartLightController(acti, name)
print(smart.light())
```

# Problem 2: Employee ID Card System

### Problem Statement

Create a class that stores employee name, ID, and department and displays ID card details.

Real-Time Use

Corporate HR systems

### üì• Sample Input

Employee Name: Rahul

Employee ID: EMP102

Department: IT

### üì§Expected Output

Employee ID Card

Name: Rahul

ID: EMP102

Department: IT


```python
class employee_card:
    def __init__(self,name,Id,department):
        self.name = name
        self.Id = Id
        self.department = department
    def status(self):
        print("employee id card details")
        print(f"name:{self.name}")
        print(f"id:{self.Id}")
        print(f"department:{self.department}")
name = input("employee name:")
employee_id = input("employee id:")
department = input("department:")
employee1 = employee_card(name,employee_id,department)
employee1.status()
```

# Problem 4: Product Price Tag Generator

### Problem Statement

Create a class that stores product name and price and prints a formatted price tag.

Real-Time Use

Retail billing systems

### üì• Sample Input

Product Name: Headphones

Price: 2499

### üì§ Expected Output

Product: Headphones

Price: ‚Çπ2499


```python
class product_price:
    def __init__(self,product,price):
        self.price = price
        self.product = product
    def show(self):
        print(f"product name:{self.product}")
        print(f"price:{self.price}")
product = input("product name:")
price = int(input("price:"))
headphones = product_price(product,price)
headphones.show()
```
