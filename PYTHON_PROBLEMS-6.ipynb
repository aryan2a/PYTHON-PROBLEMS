# Problem Statement 1:Smart Parking Lot Management System

### Design a function to manage a smart parking lot.
The system should:

Accept vehicle entry and exit logs

Calculate total parked vehicles

Identify peak parking usage

Alert if parking exceeds capacity


### Real-Time Use

Mall parking systems

Smart city infrastructure


### Hint

Function + list

Loop for counting

Conditional alert

### Sample Input

Parking Capacity: 50

Vehicle Logs: ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

### Expected Output

Currently Parked Vehicles: 3

Parking Status: Available


```python
def parking():
    parking_capacity = input("parking capacity:")
    my_list = input("vehicle logs:").split()
    count_in = 0
    count_out = 0
    for i in my_list:
        if(i.lower() == "in"):
            count_in += 1
        elif(i.lower() == "out"):
            count_out += 1
    if(count_in - count_out == parking_capacity):
        print("Parking Status: NOT AVAILABE")
    else:
        print("Parking Status: AVAILABE")

parking()
    
    
```

# Problem Statement 2:Online Food Delivery Time Estimator

### Create a function that estimates delivery time based on:

Distance (km)

Weather condition

Traffic level

Apply delays dynamically and display final ETA.

### Real-Time Use

Food delivery apps

Logistics platforms


### Hint

Function + multiple conditions

Mathematical adjustments

### Sample Input

Distance (km): 8

Traffic Level: High

Weather: Rainy

### Expected Output

Estimated Delivery Time: 55 minutes


```python
def estimated_eta(distance,traffic,weather):
    speed = 30 # km per hour
    base_time = (distance / speed)*60 # minutes
    traffic_factor = {
        "low":1.0,
        "medium":1.20,
        "high":1.30
    }

    weather_factor = {
        "low":1.0,
        "medium":1.20,
        "high":1.30
    }

    t_factor = traffic_factor.get(traffic.lower(), 1)
    w_factor = weather_factor.get(weather.lower(), 1)

    eta = base_time*t_factor*w_factor

    return round(eta)
    

distance = int(input("Distance:"))
traffic = input("traffic level:")
weather = input("weather:")
time = estimated_eta(distance,traffic,weather)
print(f"time:{time}minutes")
```


# Problem Statement 4:Cloud Server Load Classification System

Create a function to classify server load based on CPU usage readings.

### Rules:

Average CPU < 50% → Normal

50%–80% → Warning

80% → Critical

### Real-Time Use

Cloud monitoring dashboards

DevOps alerting systems


### Hint

Function + loop

Average calculation

### Sample Input

CPU Readings (%): [45, 60, 70, 85, 90]

### Expected Output

Average CPU Load: 70%

Server Status: Warning




```python
def server():
    cpu_readings = (input("cpu readings:")).split()
    cpu_readings_new = list(map(int,cpu_readings))
    average = sum(cpu_readings_new)/len(cpu_readings_new)
    print(f"average cpu load:{average}%")
    if(average < 50):
        print("server status: normal")
    elif(80 > average >= 50):
        print("server status: warning ")
    else:
        print("critical")
server()
```

# Problem Statement 5:Smart Classroom Resource Usage Monitor

Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.

### Real-Time Use

Smart classrooms

Energy optimization systems

### Hint

Function + dictionary

Conditional checks

### Sample Input

Resource Usage (hours):

{
 "Projector": 6,
 "AC": 9,
 "Lights": 4
}

### Expected Output

Overused Resources: AC

Energy Alert: Yes


```python
def smart():
    dict1 = {}
    dict1["projector"] = int(input("number of hours projector runs:"))
    dict1["AC"] = int(input("number of hours AC runs:"))
    dict1["Lights"] = int(input("number of hours lights run:"))
    for i in dict1:
        highest = "projector"
        if(dict1[i] > dict1[highest]):
            highest = i
    print(f"overused resources:{highest}")
smart()
            
            
```

# Problem Statement 6: Online Event Registration Capacity Controller

Create a function that manages event registrations by:

Tracking registrations

Preventing overbooking

Triggering waitlist mode


### Real-Time Use

Webinar platforms

Conference registration systems

### Hint

Function + loop

Capacity validation

### Sample Input

Event Capacity: 100

Registrations: 105

### Expected Output

Confirmed Registrations: 100

Waitlisted Users: 5

Registration Status: Closed



```python
def registeration():
    capacity = int(input("Event capacity:"))
    registerations = int(input("Registerations:"))
    print("confirmed registerations:",capacity)
    if(registerations == capacity):
        return "registeration status:closed"
    elif(registerations > capacity):
        print(f"waitlisted users:{registerations - capacity}")
        return "registeration status:closed"
    else:
        return"registeration status : still open"
event = registeration()
print(event)
```
