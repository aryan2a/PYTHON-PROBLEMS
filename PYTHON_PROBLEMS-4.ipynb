# Problem 1: Mobile Recharge Validation System

Problem Statement
Create a function to validate a mobile recharge request.

Rules:
Recharge amount must be ≥ ₹50

Valid plans: 199, 299, 399, 599

User should be allowed to retry if invalid

Real-Time Use
Telecom recharge apps
Payment gateways

Hint
Use function for validation,Use while loop for retry




```python
def mobile_recharge(plan):
    valid_plan = [199, 299, 399, 599]
    if(plan not in valid_plan):
              print("RETRY! PLAN NOT IN VALID PLANS")
              plan = int(input("Enter the plan for mobile recharge"))
              return mobile_recharge(plan)
    else:
        return "success" 

plan = int(input("Enter the plan for mobile recharge"))
recharge = mobile_recharge(plan)
print(recharge)
# A function should return something otherwise it will automatically return none 
# use return at the end of program or if you want to call a function otherwise it will automatically print none
# this also has o(n) time complexity
```

# Problem 2: Inventory Reorder Alert System
Problem Statement
Given a dictionary of products and stock quantities:

Print Reorder Alert if stock < 15

Print Stock OK otherwise

Real-Time Use
Warehouse systems
Retail management


Hint
Use function


Use for loop on dictionary







```python
def inventory():
    products = {"jeans":53, "cargo":34, "shirts":12 , "tshirt":10}
    for prod in products:
        if(products[prod] > 15):
            print(f"Stock OK for {prod}")
        else:
            print(f"Reorder Alert for {prod}")

inventory()
# if you dont want to write return in a function then just dont store it any variable
    
            
        

```

# Problem 3: Student Result Processing System

Problem Statement
Create a function that:
Accepts a list of student marks

Calculates average marks
Prints Pass if average ≥ 50 else Fail


Real-Time Use
Online exam systems
,Evaluation portals


Hint
Use loop to calculate total
,Return result from function







```python
def marks():
    n = int(input("enter the number of subjects:"))
    list = []
    for i in range(1,n+1):
        score = int(input(f"Enter the marks for subject {i}"))
        list.append(score)
    average = sum(list)/len(list)
    print("average marks:",average)
    if(average >= 50):
        print("Pass")
    else:
        print("Fail")
marks()
        
        
        
```

# Problem 5: Employee Attendance Eligibility Checker
Problem Statement
Create a function that:
Accepts attendance list ("P", "A")


Calculates attendance percentage


Returns Eligible if ≥ 75% else Not Eligible


Real-Time Use
HR payroll systems


Internship tracking


 Hint
Use loop to count present days


Use return statement




```python
def attendance():
    attendance_list = []
    count = 0
    for i in range(1,31):
        attend = input(f"give attendance for day{i} (answer in P or A):")
        attendance_list.append(attend)
        if(attend.lower() == "p"):
            count += 1
    attendance_percentage = (count/len(attendance_list))*100
    print("attendance percentage is:",attendance_percentage)
    if(attendance_percentage >= 75):
        return "Eligible"
    else:
        return "Not Eligible"

attendance()
    

    
```

# Problem 6: Password Strength Checker
Problem Statement
Create a function to check password strength.
Rules:
Minimum length = 8


Must contain at least one digit


Must contain at least one special character (@#$)


Real-Time Use
Login systems


Security validation


Hint
Use loop over string


Use conditions inside function



```python
def passw():
    while True:
        digits = ['1','2','3','4','5','6','7','8','9']
        special = ['@','#','$']
        count_1 = 0
        count_2 = 0
        password = input("enter the password:")
        for ch in password:
            if(ch in digits):
                count_1 += 1
            elif(ch in special):
                count_2 += 1
        if(count_1 >= 1 and count_2 >= 1 and len(password) >= 8):
            print("password is saved")
            return
        else:
            print("password must contain atleast one digit and one special character also it should have minimum length of 8")
            continue
passw()
            
            

    
```
